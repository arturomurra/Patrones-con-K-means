# Patrones con K-means

```python
diabetes_prep=diabetes.drop(['SkinThickness','Outcome'],axis=1)
```

```python
diabetes_prep.head()
```

```python
diabetes.isnull().sum()
```

```python
#k=2
version1=diabetes_prep.loc[:,["Glucose","Insulin"]]
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(version1)
labels = kmeans.predict(version1)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.scatter(diabetes_prep['Glucose'], diabetes_prep['Insulin'], c = labels)
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.show()
```

```python
df = pd.DataFrame({'labels':labels, 'Outcome':diabetes['Outcome']})
crosstab = pd.crosstab(df['labels'],df['Outcome'])
crosstab
```

```python
version2=diabetes_prep.loc[:,["BMI","DiabetesPedigreeFunction"]]
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
kmeans.fit(version2)
labels2 = kmeans.predict(version2)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.scatter(diabetes_prep['BMI'], diabetes_prep['DiabetesPedigreeFunction'], c = labels)
plt.xlabel('DiabetesPedigreeFunction')
plt.ylabel('BMI')
plt.show()
```

```python
df = pd.DataFrame({'labels2':labels2, 'Outcome':diabetes['Outcome']})
crosstab = pd.crosstab(df['labels2'],df['Outcome'])
crosstab
```

Cluster 1: 213 sin diabetes, 184 con diabetes
Cluster 0: 287 sin diabetes, 84 con diabetes

```python
iner_list = np.empty(10)
for i in range(1,10):
    kmeans2 = KMeans(n_clusters=i)
    kmeans2.fit(version1)
    iner_list[i] = kmeans2.inertia_

# iner_list = iner_list%100
# show the best number in graph
plt.plot(range(0,10), iner_list,'-')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
version1=diabetes_prep.loc[:,["Glucose","Insulin"]]
from sklearn.cluster import KMeans
kmeans_1 = KMeans(n_clusters=3)
kmeans_1.fit(version1)
labels_1 = kmeans_1.predict(version1)
list_clusters = np.unique(labels_1)
print("Número de clusters:",len(list_clusters))
print("Cluster Centers:",kmeans_1.cluster_centers_)
import matplotlib.pyplot as plt
import numpy as np
plt.scatter(diabetes_prep['Glucose'], diabetes_prep['Insulin'], c = labels_1)
plt.scatter(kmeans_1.cluster_centers_[:, 0], kmeans_1.cluster_centers_[:, 1], c='red', marker='x')
plt.xlabel('Glucose')
plt.ylabel('Insulin')
plt.show()
```

¿Crees que estos centros puedan ser representativos de los datos? ¿Por qué? Si pueden ser porque representan las tendencias de la gente.

¿Cómo obtuviste el valor de k a usar?El inicial se obtuvo porque queríamos saber si la persona era diabetica o no pero en base a las gráficas salió que debimos usar 3.

¿Los centros serían más representativos si usaras un valor más alto? ¿Más bajo? Nos hubiera salido mejor si hubieramos usado el 3 como valor de k.
